```python
x=5 #Nested if
if x>1:
  print('More than one');
  if x<100:
    print('Less than 100')
print('All done');
```

```python
x=5 #one way
print('Before 5')
if x==5:
  print('Is 5')
  print('Is still 5')
  print('Third 5')
```

```python
x=5 #multi way
if x<2:
  print('small')
elif x<10:
  print('Medium')
else:
  print('Large')
print('All done')
```

```python
astr=input('Type a number: ')
try:
  istr=int(astr)
  print('The number is,',istr)
except:
  raise ValueError('Something is wrong')
print('Done')
```

```python
n=5  #while
while n>0:
   print(n)
   n=n-1
print('Done')
```

```python
n=5 #break
while n>0:
  print(n)
  n=n-1
  if n==1:
    break
print('Done')
```

```python
n=5 #continue
while n>0:
  print(n)
  n=n-1
  if n==1:
    continue
print('Done')
```

```python
numbers=[5,4,3,2,1] #for loop listlar bilan
for n in numbers:
  print(n)
```

```python
my_dict={'a':1, 'b':2, 'c':3} #for loop dictionary bilan.
for key, value in my_dict.items():
  print(f'Key: {key}, Value:{value}')
```
