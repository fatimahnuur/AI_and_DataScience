```python
fruits=['aplle','banana','mango']#list yaratish
print(fruits)
```

```python
fruits.append('peach')#listga yangi element qo'shish
fruits
```

```python
fruits.extend(['pineapple','lemon'])#listga yangi elementlar qo'shish
fruits
```

```python
fruits.insert(1,'coconut') #listga index bo'yicha element qo'shish
fruits
```

```python
fruits.remove('peach')#listdan elementni o'chirish
fruits
```

```python
fruits.pop(0) #index bo'yicha o'chirish
fruits
```

```python
fruits.reverse() #elementlarni teskari qilish
fruits
```

```python
fruits.sort() #listdagi elementlarni tartiblash
fruits
```

```python
new_fruits=fruits[:] #Listni nusxalash
print(new_fruits)

new_fruits2=list(fruits)
print(new_fruits2) #Listni nusxalash 2-usul
```

```python
import copy #listni nusxalash 3-usul
new_fruits=copy.copy(fruits)
print(new_fruits)
```
